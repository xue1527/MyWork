Plot Daily AMSR2 Sea Ice From: 
http://www.iup.uni-bremen.de:8084/amsr2data/asi_daygrid_swath/n6250/2015/jan/
Hear take 'asi-AMSR2-n6250-20150101-v5.hdf' for plotting test.

```python
#reading AMSR2 data
from pyhdf.SD import *
import numpy as np
infile='c:\\asi-AMSR2-n6250-20150101-v5.hdf'
hf=SD(infile)
data=hf.select('ASI Ice Concentration')
ice=data[:]
data.endaccess()
#mask invalid data
ice=np.ma.masked_invalid(ice)
ice=np.ma.masked_less(ice,0)
```

```python
import matplotlib.path as mpath
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors as mcolors

def make_map():
    #===================================================
    fig=plt.figure(figsize=(6, 6))
    ax=plt.axes(projection=ccrs.NorthPolarStereo())
    box=[-180, 180, 55, 90]
    ax.set_extent(box, crs=ccrs.PlateCarree())
    #===================================================
    scale='50m'
    land = cfeature.NaturalEarthFeature('physical', 'land', scale,edgecolor='face',
                                                              facecolor=cfeature.COLORS['land'])
    ocean = cfeature.NaturalEarthFeature('physical', 'ocean', scale,edgecolor='face',
                                                             facecolor=cfeature.COLORS['water'])
    #===================================================
    ax.add_feature(ocean,facecolor='b',zorder=-1)
    ax.add_feature(land,facecolor='0.75')
    ax.coastlines(scale,linewidth=0.35) 
    #===================================================
    #clip axes by circle
    theta = np.linspace(0, 2*np.pi, 100)
    center, radius = [0.5, 0.5], 0.5
    verts = np.vstack([np.sin(theta), np.cos(theta)]).T
    circle = mpath.Path(verts * radius + center)
    ax.set_boundary(circle, transform=ax.transAxes)
    return fig,ax

#===================================================
fig,ax=make_map()
#make x,y points matching AMSR2 6250 data
dx = dy = 6250
x = np.arange(-3850000, 3750000, +dx)
y = np.arange(-5350000, 5850000, +dy)

#===================================================
#create my color map
rcol=['00','05','0A','14','1A','1F','24','2E','33','38','3D','42','4D','52','57','5C','66','6B',
          '70','75','7A','85','8A','8F','94','99','A3','A8','AD','B3','BD','C2','C7','CC','D1','DB',
          'E0','E6','EB','F5','FA','FF']
gcol=['00','05','0A','14','1A','1F','24','2E','33','38','3D','42','4D','52','57','5C','66','6B',
          '70','75','7A','85','8A','8F','94','99','A3','A8','AD','B3','BD','C2','C7','CC','D1','DB',
          'E0','E6','EB','F5','FA','FF']
bcol=['FF']*42
colors=[]
for r,g,b in zip(rcol,gcol,bcol):
    colors.append('#'+r+g+b)
my_cmap = mcolors.ListedColormap(colors)

#===================================================
#plot AMSR2 arctic sea ice concentration data
cs = ax.pcolormesh(x, y, ice, cmap=my_cmap,
                   transform=ccrs.NorthPolarStereo(central_longitude=-45))
plt.colorbar(cs,shrink=0.75,pad=0.1)

#ouput figure
fig.savefig('c:\\Ice20150101.png',dpi=150)
```
